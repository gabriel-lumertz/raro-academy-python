## **Data Frame: Array (Matriz) Bidimensioal capaz de conter qualquer tipo de dado**

```
dados = {
    'Pais': ['Belgium',  'India',  'Brazil', 'Usa', 'Japan'],
    'Capital': ['Brussels',  'New Delhi',  'Brasilia', 'Ws', 'Tokio'],
    'Populacao': [11190846, 1303171035, 207847528, 412984123, 57354123]
}
```

```
dados
```

```
# dfPais = pd.DataFrame(dados,columns=['Pais',  'Capital',  'Populacao'])
dfPais = pd.DataFrame(dados)
```

```
dfPais
```

## **Listando Dados: Comando HEAD**

```
dfPais.head(2)
```

## **Excluindo Linhas e Colunas: Comando Drop**

```
#Excluindo uma linha pelo indice
dfPais.drop(3, axis=0, inplace=True)
dfPais
```

```
#Excluindo uma coluna
dfPais.drop('Pais', axis=1, inplace=True)
dfPais
```

```
# Reconstruindo DataFrame
dfPais = pd.DataFrame(dados,columns=['Pais',  'Capital',  'Populacao'])
dfPais
```

## **Selecionando Dados: Comandos de Seleção**

```
dfPais[:]
```

```
dfPais[1:3]
```

```
# Selecionado dados por Linha
# iloc[[linha],[coluna]]
# iloc[[linha]]
dfPais.iloc[[1,3],[0,2]]
```

```
dfPais['Capital']
```

```
# Filtrado os Dados
dfPais[dfPais['Pais'] == 'Brazil']
```

```
#População Maior que 60.000.0000
dfPais[dfPais['Populacao'] > 60000000]
```

```
dfPais[(dfPais['Populacao'] > 60000000) & (dfPais['Populacao'] < 500000000)]
```

```
dfPais.plot.bar('Capital', 'Populacao')
```

```
dfPais.plot.bar('Pais', 'Populacao')
```

## **Exibindo informações estatísticas dos Dados**

```
# Reconstruindo o DataFrame
dfPais = pd.DataFrame(dados,columns=['Pais',  'Capital',  'Populacao'])
dfPais
```

```
# Informações gerais do DataSet
dfPais.describe()
```

```
# Median exibe o dado mediado (do meio) e Mean exibe a média
print(dfPais.median(), '\n\n', dfPais.mean())
```

```
# Max exibe o maior valor e Min exibe o menor valor
print(dfPais.max(), '\n\n', dfPais.min())
```
